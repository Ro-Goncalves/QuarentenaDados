```python
nota_medias_por_filme = avaliacoes.groupby("filmeId")['nota'].mean()
```

# Encontrar os filmes que não possuem avaliação

```python
filmes_com_media = filmes.join(nota_medias_por_filme, on='filmeId')
filmes_com_media.head()
```
